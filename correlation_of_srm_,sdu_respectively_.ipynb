#srm data correaltion 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
correlation_matrix = srm.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sdu = pd.read_csv("/content/sdu0001-wmu4g-_7_.csv")
correlation_matrix = sdu.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
